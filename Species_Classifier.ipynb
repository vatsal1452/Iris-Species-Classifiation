Making Machine Learning Model Using Tree

```python
# Splitting Data Frame into test and train data
```

```python
y=df['Species']
```

```python
y
```

```python
x=df[['Sepal.Length','Petal.Length']]
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_train,y_train
```

```python
#importing tree module from sklearn
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
tree=DecisionTreeClassifier(random_state=0)
tcl=tree.fit(x_train,y_train)
```

```python
tcl.predict(x_test)
```

```python
x_test.shape
```

```python
plot_tree(tcl)
```

```python
c=tcl.predict(x_test)
```

```python
c.shape
```

```python
y_test.shape
```

```python
print(y_test)
```

```python
y_test
```

```python
# Creating Confusion Matrix
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,c)
```
